### Данные

Для этого ДЗ мы будем использовать датасет треков со стримингового сервиса Spotify

**Описание данных**

- **track_id:** The Spotify ID for the track
- **artists:** The artists' names who performed the track. If there is more than one artist, they are separated by a ;
- **album_name:** The album name in which the track appears
- **track_name:** Name of the track
- **popularity:** The popularity of a track is a value between 0 and 100, with 100 being the most popular. The popularity is calculated by algorithm and is based, in the most part, on the total number of plays the track has had and how recent those plays are. Generally speaking, songs that are being played a lot now will have a higher popularity than songs that were played a lot in the past. Duplicate tracks (e.g. the same track from a single and an album) are rated independently. Artist and album popularity is derived mathematically from track popularity.
- **duration_ms:** The track length in milliseconds
- **explicit:** Whether or not the track has explicit lyrics (true = yes it does; false = no it does not OR unknown)
- **danceability:** Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable
- **key:** The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1
- **loudness:** The overall loudness of a track in decibels (dB)
- **mode:** Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0
- **speechiness:** Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks
- **acousticness:** A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic
- **instrumentalness:** Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content
- **liveness:** Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live
- **valence:** A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry)
- **tempo:** The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration
- **time_signature:** An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). The time signature ranges from 3 to 7 indicating time signatures of 3/4, to 7/4.
- **track_genre:** The genre in which the track belongs

**Целевая переменная**
- **energy:** Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale

```python
# Категоральные не закодированные признаки. Мы их закодируем.
categorical_columns = ['explicit', 'track_genre']
# Строковые "бесполезные" признаки.
string_columns = ['artists', 'album_name', 'track_name']
# Категоральные уже закодированные признаки.
categorical_columns_coded = ['key', 'mode', 'time_signature']

# Используем LabelEncoder.
label_encoder = LabelEncoder()

# Очень хочется сохранить соответсвия.
map_for_categorical_columns = {}

# Применим LabelEncoder к каждой коллонке, которую посчитали категоральной.
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])
    map_for_categorical_columns[column] = label_encoder.classes_

Y = data['energy']
X = data.drop(['energy'], axis=1)

data
```

```python
# Делим данные на тренировочные и тестовые с заддаными пропорцией и стейтом.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=7)
```

#### 2. [0.75 балла] Обучите модели на train'е, исключив категориальные признаки, используя библиотеку StatsModels и примените ее к test'у; используйте $RMSE$ и $R ^ 2$ в качестве метрики качества. Попробуйте также применить реализации линейной регрессии из sklearn:

* [`LinearRegression`](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html);
* [`Ridge`](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html) with $\alpha = 0.03$;
* [`Lasso`](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html) with $\alpha = 0.05$
* [`ElasticNet`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.ElasticNet.html) with $\alpha = 0.01$, $l_{1}$_$ratio = 0.4$

Не забывайте скейлить данные с помощью StandardScaler перед обучением моделей! 

```python
def evaluate(Y_true, Y_pred, now_model="None", need_print=True):
    """
    Функция для рассчета RMSE и R^2.
    ---
    Y_true: np.array of shape (l, d) 
            Истиные значения цели.
    Y_pred: np.array of shape (l, d)
            Предсказанные значения цели.
    now_model: str
            Название модели.
    need_print: bool
            Нужно ли печатать информацию.
    ---
    output: float, float 
    """
    rmse = np.sqrt(mean_squared_error(Y_true, Y_pred))
    r_2 = r2_score(Y_true, Y_pred)
    if need_print:
        print(f"Модель {now_model}\nRMSE: {rmse}\tR^2: {r_2}\n\n")
    return rmse, r_2

def info_features(set):
    """
    Функция для показа информации о признаках.
    ---
    set: data 
         Набор данных.
    ---
    output: None 
    """
    print(f"Работаем с {len(set.columns)} признаками:\n")
    for feature in list(set.columns):
        print(f"\t{feature}")
    print("\n")


# Исключаем категоральные признаки и названия треков, альбомов и имена исполнителей для train X и test X.
X_train_droped_categorical = X_train.drop(categorical_columns + string_columns + categorical_columns_coded, axis=1)
X_test_droped_categorical = X_test.drop(categorical_columns + string_columns + categorical_columns_coded, axis=1)

# Получаем данные о признаках.
info_features(X_train_droped_categorical)

# Используем StandardScaler.
scaler = StandardScaler()
# fit_transform для тренировочных данных, а transform для тестовых.
X_train_scaled = scaler.fit_transform(X_train_droped_categorical)
X_test_scaled = scaler.transform(X_test_droped_categorical)

# StatsModels: Линейная регрессия.
X_train_stats = np.column_stack((np.ones(X_train_scaled.shape[0]), X_train_scaled))
sm_lr = sm.OLS(Y_train, X_train_stats).fit()
X_test_stats = np.column_stack((np.ones(X_test_scaled.shape[0]), X_test_scaled))
Y_pred_sm = sm_lr.predict(X_test_stats)

# StatsModels: Ridge.
sm_ridge_stats = sm.OLS(Y_train, X_train_stats).fit_regularized(alpha=0.03, L1_wt=0.0)
Y_pred_ridge_stats = sm_ridge_stats.predict(X_test_stats)

# StatsModels: Lasso.
sm_lasso_stats = sm.OLS(Y_train, X_train_stats).fit_regularized(alpha=0.05, L1_wt=1.0)
Y_pred_lasso_stats = sm_lasso_stats.predict(X_test_stats)

# StatsModels: ElasticNet.
sm_elastic_net_stats = sm.OLS(Y_train, X_train_stats).fit_regularized(alpha=0.01, L1_wt=0.4)
Y_pred_elastic_net_stats = sm_elastic_net_stats.predict(X_test_stats)

# sklearn: Линейная регрессия.
lr = LinearRegression()
lr.fit(X_train_scaled, Y_train)
Y_pred_lr = lr.predict(X_test_scaled)

# sklearn: Ridge.
ridge = Ridge(alpha=0.03)
ridge.fit(X_train_scaled, Y_train)
Y_pred_ridge = ridge.predict(X_test_scaled)

# sklearn: Lasso.
lasso = Lasso(alpha=0.05)
lasso.fit(X_train_scaled, Y_train)
Y_pred_lasso = lasso.predict(X_test_scaled)

# sklearn: ElasticNet.
elastic_net = ElasticNet(alpha=0.01, l1_ratio=0.4)
elastic_net.fit(X_train_scaled, Y_train)
Y_pred_elastic_net = elastic_net.predict(X_test_scaled)

# Создаём dict с названием модели и ее предсказанием.
models_1 = {"StatsModels: Линейная регрессия": [Y_pred_sm],
         "StatsModels: Ridge": [Y_pred_ridge_stats],
         "StatsModels: Lasso" : [Y_pred_lasso_stats],
         "StatsModels: ElasticNet" : [Y_pred_elastic_net_stats],
         "sklearn: Линейная регрессия": [Y_pred_lr],
         "sklearn: Ridge": [Y_pred_ridge],
         "sklearn: Lasso": [Y_pred_lasso],
         "sklearn: ElasticNet": [Y_pred_elastic_net]}

# А теперь производим оценку и выводим её на экран.
for now_model, Y_pred_value in models_1.items():
    rmse, r_2 = evaluate(Y_test, Y_pred_value[0], now_model)
    models_1[now_model].append(rmse)
    models_1[now_model].append(r_2)
```

```python
from sklearn.feature_selection import f_regression

# Воспользуемся немножко семинарским опытом...
def show_weights(features, weights, scales, name):
    """
    Функция для показа весов модели.
    ---
    features: array
              Набор признаков.
    weights: array
              Набор весов.
    scales: array
              Набор scales.
    name: str
              Имя модели.
    ---
    output: None
    """
    plt.figure(figsize=(8, 6))
    s_weights = sorted(zip(weights, features, scales), reverse=True)
    sns.barplot(y=[x[1] for x in s_weights], x=[x[0] for x in s_weights])
    plt.title(name)
    plt.xlabel("Вес")
    plt.ylabel("Фича")

# Агрегируем данные по именам и моделям.
names = ["StatsModels: Линейная регрессия (без категоральных)",
         "StatsModels: Ridge (без категоральных)",
         "StatsModels: Lasso (без категоральных)",
         "StatsModels: ElasticNet (без категоральных)",
         "sklearn: Линейная регрессия (без категоральных)",
         "sklearn: Ridge (без категоральных)",
         "sklearn: Lasso (без категоральных)",
         "sklearn: ElasticNet (без категоральных)",
         "StatsModels: Линейная регрессия",
         "StatsModels: Ridge",
         "StatsModels: Lasso",
         "StatsModels: ElasticNet",
         "sklearn: Линейная регрессия",
         "sklearn: Ridge",
         "sklearn: Lasso",
         "sklearn: ElasticNet"]
models_all = [sm_lr, sm_ridge_stats, sm_lasso_stats, sm_elastic_net_stats, lr, ridge, lasso, elastic_net, sm_lr_2, sm_ridge_stats_2, sm_lasso_stats_2, sm_elastic_net_stats_2, lr_2, ridge_2, lasso_2, elastic_net_2]

# Выведем кастомную статистику по каждой из 10 моделей, включая графики весов.
for model_index in range(0, len(models_all)):
    # Печатаем имя моделию.
    print(f"\n\n\t\t\t{names[model_index]}")
    if model_index not in [0, 1, 2, 3, 8, 9, 10, 11]:
        # Для sklearn моделей
        if model_index < len(models_all) / 2:
            # Для sklearn моделей обученных без категориальных признаков.
            scales = pd.Series(data=X_train_scaled.std(axis=0), index=X_train_droped_categorical.columns)
            show_weights(X_train_droped_categorical.columns, models_all[model_index].coef_, scales, names[model_index])
            print("Коэффиценты:\n")
            for i, coefficient in enumerate(models_all[model_index].coef_):
                print(f"{X_train_droped_categorical.columns[i]}:\t{coefficient if coefficient != 0.0 else 'Нулевой!'}")
            print(f"\nПересечение: {models_all[model_index].intercept_}\n")
            f_statistic, p_values = f_regression(X_train_scaled, Y_train)
            for i, (f_stat, p_value) in enumerate(zip(f_statistic, p_values)):
                print(f"{X_train_droped_categorical.columns[i]}:\nF-статистика = {f_stat}\tp-значение = {p_value}")
        else:
            # Для sklearn моделей обученных с категориальными признаками.
            scales = pd.Series(data=X_train_scaled_2.std(axis=0), index=X_train_drop.columns)
            show_weights(X_train_drop.columns, models_all[model_index].coef_, scales, names[model_index])
            print("Коэффиценты:\n")
            for i, coefficient in enumerate(models_all[model_index].coef_):
                print(f"{X_train_drop.columns[i]}:\t{coefficient if coefficient != 0.0 else 'Нулевой!'}")
            print(f"\nПересечение: {models_all[model_index].intercept_}\n")
            f_statistic, p_values = f_regression(X_train_scaled_2, Y_train)
            print("F-статистика и p-значения для каждого предиктора:")
            for i, (f_stat, p_value) in enumerate(zip(f_statistic, p_values)):
                print(f"{X_train_drop.columns[i]}:\nF-статистика = {f_stat}, p-значение = {p_value}")
    elif model_index in [0, 8]:
        # Для OLS моделей с имплементированным summary.
        print(models_all[model_index].summary())
    else:
        # Для OLS моделей с НЕ имплементированным summary.
        print("Коэффиценты:\n")
        num = 0
        for i in models_all[model_index].params:
            if num == 0:
                print(f"const:\t{i if i != 0 else 'Нулевой!'}")
            elif model_index < len(models_all) / 2:
                print(f"{X_train_droped_categorical.columns[num - 1]}:\t{i if i != 0.0 else 'Нулевой!'}")
            else:
                print(f"{X_train_drop.columns[num - 1]}:\t{i if i != 0.0 else 'Нулевой!'}")
            num += 1
```

### Развернутый ответ:
Давайте сначала поймем какие веса получились нулевыми:\

Похожую ситуацию можно наблюдать с **Ridge**.\
**Lasso**, в свою очередь, окзывается интересна своей "неадекватностью", по сравнению с другими моделемя она является незначимой и имееет наихудшие показатели, не зануляет веса у таких признаков, как: *loudness* и acousticness, все остальные веса равны 0. Конечно, большой вес у *loudness* - неплохая вещь, обеспечивающая приемлемые метрики, но в сравнении, все плохо.\
**Elastic Net** зануляет все веса, связанные с категоральными признаками *track_genre* и *explicit*. А также зануляет key и приводит близко к нулю *mode* и *popularity*, что я считаю вполне нормальным.\
\
В принципе, все модели, с точки зрения наших изначальных метрик оказались значимыми, вопросы оставляет только Lasso, она показывает себя куда хуже в сравнении и рассмотрев веса, ставновится понятно почему.

## Градиентный спуск

#### 7. [3.5 балла] Имплементируйте  Ridge регрессию для MSE loss, обученную на градиентом спуске.


Все вычисления должны быть векторизованы, а циклы Python можно использовать только для итераций градиентного спуска. В качестве критерия остановки необходимо использовать (одновременно):

* проверка абсолютной нормы разницы весов на двух соседних итерациях (например, меньше некоторого малого числа порядка $10^{-6}$, заданного параметром `tolerance`);

* достижение максимального количества итераций (например, 10000, заданного параметром `max_iter`).

Вам необходимо выполнить:

* Полный градиентный спуск:

$$
w_{k + 1} = w_{k} - \eta_{k} \nabla_{w} Q(w_{k}).
$$

* Стохастический градиентный спуск:

$$
w_{k + 1} = w_{k} - \eta_{k} \nabla_{w} q_{i_{k}}(w_{k}).
$$

$\nabla_{w} q_{i_{k}}(w_{k}) \, $ является оценкой градиента по набору объектов, выбранных случайным образом.

* Momentum method:

$$
h_0 = 0, \\
h_{k + 1} = \alpha h_{k} + \eta_k \nabla_{w} Q(w_{k}), \\
w_{k + 1} = w_{k} - h_{k + 1}.
$$

* Adagrad method:

$$
G_0 = 0, \\
G_{k + 1} = G_{k} + (\nabla_{w} Q(w_{k+1}))^2, \\
w_{k + 1} = w_{k} - \eta * \frac{\nabla_{w} Q(w_{k+1})}{\sqrt{G_{k+1} + \epsilon}}.
$$

Чтобы убедиться, что процесс оптимизации действительно выполняется, мы будем использовать атрибут класса `loss_history`. После вызова метода fit он должен содержать значения функции потерь для всех итераций, начиная с первой (до первого шага по антиградиенту).


Вам нужно инициализировать веса случайным вектором из нормального распределения. Ниже приведен шаблон, который должен содержать код, реализующий все варианты моделей.

![](https://i.redd.it/9euoe42zwj641.jpg)

```python
from sklearn.base import BaseEstimator

def print_itaration_data(iteration, max_iter, loss, each_iter=10):
    """
    Функция для отображения функции потерь на каждой each_iter итерации.
    ---
    iteration: int
               текущая итерация.
    max_iter: int
              Набор весов.
    loss: float
          текущая функция потерь.
    each_iter: int
               каждую какую итерацию отображать.
    ---
    output: None
    """
    if iteration % each_iter == 0:
        print(f"Итерация {iteration} из {max_iter}\tLoss: {loss}")
    else:
        return

methods = ['FullGradient', 'Momentum', 'Adagrad', 'StochasticGradient']

class LinReg(BaseEstimator):
    def __init__(self, delta=1.0, gd_type='Momentum', 
                 tolerance=1e-4, max_iter=1000, w0=None, eta=1e-2, alpha=1e-3, reg_cf=1.0, epsilon=1e-8, print_iter=False):
        """
        gd_type: str
            'GradientDescent', 'StochasticDescent', 'Momentum', 'Adagrad'
        delta: float
            proportion of object in a batch (for stochastic GD)
        tolerance: float
            for stopping gradient descent
        max_iter: int
            maximum number of steps in gradient descent
        w0: np.array of shape (d)
            init weights
        eta: float
            learning rate
        alpha: float
            momentum coefficient
        reg_cf: float
            regularization coefficient
        epsilon: float
            epsilon
        print_iter: bool
            print of iteration info is needed
        """
        
        self.delta = delta
        self.gd_type = gd_type
        self.tolerance = tolerance
        self.max_iter = max_iter
        self.w0 = w0
        self.alpha = alpha
        self.w = None
        self.eta = eta
        self.loss_history = None
        self.reg_cf = reg_cf
        self.print_iter = print_iter
        self.epsilon = epsilon
    
    def fit(self, X, y):
        """
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: self
        """
        # Готовим истории функции потерь.
        self.loss_history = []
        # Заполняем весами из нормального распределения, если не заданы начальные.
        self.w = np.random.normal(0, 1, X.shape[1]) if self.w0 is None else self.w0
        
        # Используем тот тип градиента, который был задан.
        if self.gd_type == 'FullGradient':
            self._full_gradient_descent(X, y)
        elif self.gd_type == 'StochasticGradient':
            self._stochastic_gradient_descent(X, y)
        elif self.gd_type == 'Momentum':
            self._momentum(X, y)
        elif self.gd_type == 'Adagrad':
            self._adagrad(X, y)
        else:
            raise Exception('Method is not implemented!')
            
        return self
    
    def predict(self, X):
        """
        Метод для вычисления градиента.
        ---
        X: np.array of shape (l, d)
        ---
        output: np.array of shape (l)
        """
        if self.w is None:
            raise Exception('Not trained yet')
            
        # Возвращаем результат с помощью домножения на веса.
        return np.dot(X, self.w)
    
    def calc_gradient(self, X, y):
        """
        Метод для вычисления градиента.
        ---
        X: np.array of shape (l, d) (l can be equal to 1 if stochastic)
        y: np.array of shape (l)
        ---
        output: np.array of shape (d)
        """
        if X.ndim == 1:
            # Градиент для стохастического.
            gradient = 2 * X * (np.dot(X, self.w) - y) + 2 * self.reg_cf * self.w
        else:
            # Градиент в остальных случаях.
            gradient = 2 * np.dot(X.T, (np.dot(X, self.w) - y)) / X.shape[0] + 2 * self.reg_cf * self.w
            
        return gradient

    
    def calc_loss(self, X, y):
        """
        Метод для вычисления функции потерь.
        ---
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: float 
        """ 
        # Получаем предсказание.
        predict = np.dot(X, self.w)
        # Считаем функцию потерь.
        loss = ((predict - y) ** 2).mean() + self.reg_cf * np.sum(self.w**2)
        
        return loss

    
    def _full_gradient_descent(self, X, y):
        """
        Метод полного градиентного спуска.
        ---
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: None 
        """ 
        for iteration in range(self.max_iter):
            gradient = self.calc_gradient(X, y)
            self.w -= self.eta * gradient
    
            loss = self.calc_loss(X, y)
            self.loss_history.append(loss)
            
            if self.print_iter:
                print_itaration_data(iteration, self.max_iter, loss)
            
            if np.linalg.norm(gradient) < self.tolerance:
                break

    def _stochastic_gradient_descent(self, X, y):
        """
        Метод стохастического градиентного спуска.
        ---
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: None 
        """ 
        y = np.array(y)
        prev_grad = np.zeros(X.shape[1])
        
        for iteration in range(self.max_iter):
            for _ in range(X.shape[0]):
                random_i = np.random.randint(0, X.shape[0])
                gradient = self.calc_gradient(X[random_i], y[random_i])
                prev_grad = self.alpha * prev_grad + gradient
                self.w -= self.eta * prev_grad
                
            loss = self.calc_loss(X, y)
            self.loss_history.append(loss)
            
            if self.print_iter:
                print_itaration_data(iteration, self.max_iter, loss)
            
            if np.linalg.norm(gradient) < self.tolerance:
                break
            

    def _momentum(self, X, y):
        """
        Метод momentum.
        ---
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: None 
        """ 
        prev_velocity = np.zeros(self.w.shape)
        
        for iteration in range(self.max_iter):
            gradient = self.calc_gradient(X, y)
            velocity = self.alpha * prev_velocity + self.eta * gradient
            self.w -= velocity
            
            loss = self.calc_loss(X, y)
            self.loss_history.append(loss)
            
            if self.print_iter:
                print_itaration_data(iteration, self.max_iter, loss)
            
            if np.linalg.norm(velocity) < self.tolerance:
                break
                
            prev_velocity = velocity

    def _adagrad(self, X, y):
        """
        Метод adagrad.
        ---
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: None 
        """ 
        g = np.zeros(self.w.shape)
        
        for iteration in range(self.max_iter):
            gradient = self.calc_gradient(X, y)
            g += gradient ** 2
            self.w -= self.eta / (np.sqrt(g + self.epsilon)) * gradient

            loss = self.calc_loss(X, y)
            self.loss_history.append(loss)

            if self.print_iter:
                print_itaration_data(iteration, self.max_iter, loss)
            
            if np.linalg.norm(gradient) < self.tolerance:
                break
```
